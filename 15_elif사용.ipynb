```python
# elif 사용하기

# if 조건식:
#     코드1
# elif 조건식:
#     코드2

# if 조건식:
#     코드1
# else:
#     코드2

x = 20

if x == 10:
    print('This is 10')
elif x == 20:
    print('This is 20')
```

```python
# if, elif, else 모두 사용허기

# if 조건식:
#     코드1
# elif 조건식:
#     코드2
# else:
#      코드3

x = 30

if x == 10:
    print('this is 10')
elif x == 20:
    print('this is 20')
else:
    print('no 10 no 20')
```

```python
# 음료수 자판기 만들기

button = int(input())

if button == 1:
    print('coke')
elif button == 2:
    print('sprite')
elif button == 3:
    print('fanta')
else:
    print('error')
```
