```python
students = {
    "김철수": [90, 80, 70],
    "이영희": [100, 95, 90],
    "박민수": [60, 70, 65]
}

print("===== 학생 성적 결과 =====")
total_sum = 0
total_count = 0

for name, scores in students.items():
    avg = sum(scores) / len(scores)
    print(f"이름: {name}")
    print(f"점수: {scores}")
    print(f"평균: {avg}")
    print()
    total_sum += sum(scores)
    total_count += len(scores)

print("===== 전체 결과 =====")
print(f"전체 학생 수: {len(students)}명")
print(f"전체 평균: {total_sum / total_count}")
```
