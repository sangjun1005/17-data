```python
# 딕셔너리 만들기

# 딕셔너리 = {
#     키1:값1
#     키2:값2
#     ...
# }

lux = {
    "health":490,
    "mana":334,
    "melee":550,
    "armor":18.71
}
print(lux)
```

```python
# 키 이름이 중복이 되면?

lux = {
    "health":490,
    "health":800,
    "mana":334,
    "melee":550,
    "armor":18.71
}
print(lux)
```

```python
# 딕셔너리 키의 자료형

x = {
    100:"hundred",
    False:0,
    3.5:[3.5,3.5]
}
print(x)
```

```python
x = {
    [10,20]:100
}
print(x)
```

```python
# 빈 딕셔너리 만들기

x = {}
y = dict()

print(x)
print(y)
print(type(x))
print(type(y))
```

```python
# dict로 딕셔너리 만들기

dict(
    health=490,
    mana=344,
    melee=550,
    armor=18.72
)
```

```python
lux=dict(zip(
    ['health','mana','melee','armor'],
    [490, 334, 550, 18.72]
))
print(lux)
```
