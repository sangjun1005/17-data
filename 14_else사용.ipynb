```python
# else와 들여쓰기

x = 5

if x ==10:
    print('10임 ㅋㅋ')
else:
    print('x에 들어있는 숫자는')
    print('10이 아님 ㅗ')
```

```python
# if 조건문에 숫자 지정하기

if 0:
    print('ㄹㅇ')
else:
    print('구라')

if 1:
    print('ㄹㅇ')
else:
    print('구라')

if 0x1f:
    print('ㄹㅇ')
else:
    print('구라')

if 0b1000:
    print('ㄹㅇ')
else:
    print('구라')

if 13.5:
    print('ㄹㅇ')
else:
    print('구라')
```

```python
# if 조건문에 문자열 지정하기

if 'hello':
    print('real')
else:
    print('fake')

if ' ':
    print('real')
else:
    print('fake')

if '':
    print('real')
else:
    print('fake')
```

```python
if not 0:
    print('real')
if not None:
    print('real')
if not '':
    print('real')
```

```python
# 조건식을 여러개 지정하기

x = 10
y = 20

if x ==10 and y ==20:
    print('real')
else:
    print('fake')
```
