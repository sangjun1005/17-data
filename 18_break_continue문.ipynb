```python
# while에서 break로 반복문 끝내기

i = 0
while True:
    print(i)

    i += 1
    
    if i == 10:
        break
```

```python
# for에서 break로 반복문 끝내기

for i in range(10000):
    print(i)
    if i == 10:
        break
```

```python
# continue로 코드 실행 건너뛰기

for i in range(10):
    if i % 2 == 0:
        continue

    print(i)
```

```python
i = 0
while i < 10:
    i += 1

    if i % 2 == 0:
        continue

    print(i)
```

```python
# 입력한 횟수대로 반복하기

count = int(input('반복할 횟수를 입력하세요.: '))

i = 0

while True:
    

    i += 1
    print(i)

    if i == count:
        break
```

```python
start, stop = map(int, input().split())
```

```python
print(start)
print(stop)
```

```python
i = start

while True:

    if i == stop + 1:
        break
    if i % 10 == 3:
        i += 1
        continue

    print(i, end=" ")
    i += 1
```
