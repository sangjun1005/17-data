```python
# 1부터 100까지 출력

for i in range(1,101):
    print(i)
```

```python
# 3의 배수일 때와 5의 배수일 때 처리하기

for i in range(1,101):

    if i % 3 == 0:
        print('Fizz')
    elif i % 5 == 0:
        print('Buzz')
    else:
        print(i)
```

```python
# 3과 5의 공배수 처리하기

for i in range(1,101):

    if i % 3 == 0 and i % 5 == 0:
        print('FizzBuzz')
    elif i % 3 == 0:
        print('Fizz')
    elif i % 5 == 0:
        print('Buzz')
    
    else:
        print(i)
```
