```python
# 리스트 표현식 사용하기
# [식 for 변수 in 리스트]
# list(식 for 변수 in 리스트)

a = [i for i in range(10)]
print(a)
```

```python
b = list(i for i in range(10))
print(b)
```

```python
c = [i + 5 for i in range(10)]
print(c)
```

```python
# 리스트 표현식에서 if 조건문 사용하기

# [식(변수) for 변수 in 리스트 if 조건식]
# list(식(변수) for 변수 in 리스트 if 조건식)

a = [i for i in range(10) if i % 2 == 0]
print(a)
```

```python
a = []
for i in range(10):
    if i % 2 == 0:
        a.append(i)

print(a)
```

```python
b = [i + 5 for i in range(10) if i % 2 == 1]
print(b)
```

```python
# for 반복문과 if 조건문을 여러번 사용하기

# a = [i*j for i in range(2,10) for j in range(1,10)]

b = []

for i in range(2,10):
    for j in range(1,10):
        b.append(i*j)

print(b)
```

```python
# 리스트에 map 사용하기

# list(map(함수,리스트))
# tuple(map(함수,튜플))

a = [1.2, 2.5, 3.7, 4.6]

for i in range(len(a)):
    a[i] = int(a[i])

print(a)
```

```python
a = [1.2, 2.5, 3.7, 4.6]
a = list(map(int, a))
print(a)
```

```python
# 튜플 표현식 사용하기

# tuple(식 for 변수 in 리스트 if 조건식)

a = tuple(i for i in range(10) if i % 2 == 0)
print(a)
```
