```python
a, b, c, d = map(int,input().split())

if a > 100 or b>100 or c>100 or d>100:
    print("error")

elif (a+b+c+d) / 4 >= 80:
    print('pass')
elif (a+b+c+d) / 4 < 80:
    print('default')


```
