```python
# 사람 클래스로 학생 클래스 만들기

# class 기반클래스이름:
#     코드


# class 파생클래스이름(기반클래스이름):
#     코드

class Person:
    def greeting(self):
        print('hello')

class Student(Person):
    def study(self):
        print('study')


james = Student()
james.greeting()
james.study()
```

```python
# 상속관계와 포함관계 알아보기

class Person:
    def greeting(self):
        print('hello')

class Personlist:
    def __init__(self):
        self.person_list = []

    def append_person(self,person):
        self.person_list.append(person)
```

```python
# 기반 클래스와 속성 사용하기

class Person:
    def __init__(self):
        print('Person __init__')
        self.hello = 'HI'

class Student(Person):
    def __init__(self):
        print('Student __init__')
        self.school = '파이썬 코딩 도장'

james = Student()
```

```python
class Person:
    def __init__(self):
        print('Person __init__')
        self.hello = 'HI'

class Student(Person):
    def __init__(self):
        print('Student __init__')
        super().__init__()
        self.school = '파이썬 코딩 도장'

james = Student()
print(james.school)
print(james.hello)
```

```python
## 기반 클래스를 초기화 하지 않아도 되는 경우

class Person:
    def __init__(self):
        print('Person __init__')
        self.hello = 'HI'

class Student(Person):
    pass

james = Student()
print(james.hello)
```

```python
# 메서드 오버라이딩 사용하기

class Person:
    def greeting(self):
        print('HI')

class Student(Person):
    def greeting(self):
        print("hi i'm coding master.")

james = Student()
james.greeting()
```

```python
class Person:
    def greeting(self):
        print('HI')

class Student(Person):
    def greeting(self):
        super().greeting()
        print("hi i'm coding master.")

james = Student()
james.greeting()
```

```python
# 다중상속 사용하기

class Person:
    def greeting(self):
        print('HI')

class University:
    def manage_credit(self):
        print('MANAGE CREDIT')

class Undergraduate(Person, University):
    def study(self):
        print('STUDY')

james = Undergraduate()
james.greeting()
james.manage_credit()
james.study()
```
