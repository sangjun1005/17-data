```python
# 문자열 조작하기
## 문자열 바꾸기

'hello, world'.replace('world','python')
```

```python
## 문자바꾸기

table = str.maketrans('aeiou','12345')
'apple'.translate(table)
```

```python
## 문자열 분리하기

text = 'apple pear grape pineapple orange'
text.split()
```

```python
text = 'apple, pear, grape, pineapple, orange'
text.split(', ')
```

```python
## 구분자 문자열과 문자열 리스트 연결하기

' '.join(['apple', 'pear', 'grape', 'pineapple', 'orange'])
```

```python
','.join(['apple', 'pear', 'grape', 'pineapple', 'orange'])
```

```python
## 소문자를 대문자로, 대문자를 소문자로

'python'.upper()
```

```python
'PYTHON'.lower()
```

```python
# 공백 삭제하기

a = '   python   '

print(a.rstrip())
print(a.lstrip())
print(a.strip())
```

```python
# 특정 문자 삭제하기

a = ' ,. python z '

print(a.rstrip(', .'))
print(a.lstrip(', .'))
print(a.strip(', .'))
```

```python
# 문자열 정렬하기

print('python'.rjust(10))
print('python'.ljust(10))
print('python'.center(10))
```

```python
# 메서드 체이닝

'python'.rjust(10).upper().strip()
```

```python
# 문자열 왼쪽에 0 채우기

print('35'.zfill(4))
print('3.5'.zfill(6))
print('035'.zfill(3))
print('035'.zfill(2))
```

```python
# 문자열 위치 찾기

print('apple pineapple'.find('a'))
print('apple pineapple'.rfind('a'))

print('apple pineapple'.index('a'))
print('apple pineapple'.rindex('a'))
```

```python
# 문자열 서식 지정자와 포매팅 사용하기

## 서식 지정자로 문자열 넣기

'I am %s.' % 'james'
```

```python
name = 'kwon'

'I am %s.' % name
```

```python
## 서식 지정자로 숫자 넣기

'I am %d years old.' % 27
```

```python
## 서식 지정자로 소수점 표현하기

print('이거 확률 %f 정도 돼' % 72.4)

print('이거 확률 %.1f 정도 돼' % 72.4)
```

```python
## 서식 지정자로 문자열 정렬하기

print('%10s' % 'python')
print('%-10s' % 'python')
```

```python
## 서식 지정자로 문자열 안에 값 여러개 넣기

'Today is %d %s.' % (3, 'July')
```

```python
## format 매서드 사용하기

print('hello {0}'.format('world'))
print('hello {0}'.format(100))
print('hello {0}{2}{1}'.format('100 ', 'python ', 'world '))

print('hello, {a} {b}'.format(a= 'python', b='world'))
```
