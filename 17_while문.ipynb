```python
# while 반복문 사용하기

# 초기식
# while 조건식:
#     반복할 코드
#     변화식

i = 0
while i < 10:
    print('hello world')
    i += 1 #i = i + 1
```

```python
for _ in range(10):
    print('hello world')
```

```python
# 초기값을 1부터 시작하기

i = 1
while i <= 10:
    print('hello world')
    i += 1
```

```python
i = 10
while i > 0:
    print('hello world',i)
    i -= 1
```
